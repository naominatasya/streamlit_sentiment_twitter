```python
kamus_alay_df = pd.read_excel(r'/content/drive/MyDrive/kamus_alay.xlsx',
                              names=['slang', 'formal'])  # Tidak perlu menggunakan encoding


# Menghapus baris dengan nilai NaN pada kolom 'formal'
kamus_alay_df = kamus_alay_df.dropna(subset=['formal'])

# Mengonversi menjadi kamus
kamus_alay_dict = dict(zip(kamus_alay_df['slang'], kamus_alay_df['formal']))
```

```python
kamus_alay_df
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
stopwords_df = pd.read_excel(r'/content/drive/MyDrive/stopwords_twitter.xlsx', header=None)

additional_stopwords = stopwords_df[0].tolist()  # Mengambil daftar stopwords dari kolom pertama

stop_words_id = set(stopwords.words('indonesian'))

# Menambahkan stopwords dari CSV ke daftar stopwords
stop_words_id.update(additional_stopwords)
```

```python

# Fungsi untuk mengganti kata slang dengan kata baku menggunakan kamus alay
def replace_slang_with_formal(text):
    # Pisahkan kata-kata berdasarkan spasi
    words = text.split()
    # Periksa apakah kata tersebut ada dalam kamus, jika ada, ganti dengan kata formal
    words = [kamus_alay_dict.get(word.lower(), word) for word in words]
    return ' '.join(words)

# Fungsi lengkap pembersihan teks
def clean_text(text):
    # Case folding: Mengubah teks menjadi lowercase
    text = text.lower()

    # Menghapus URL
    text = re.sub(r'http\S+', '', text)

    # Menghapus mention (contoh: @username) dan hashtag (contoh: #hashtag)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#\w+', '', text)

    # Menghapus karakter non-alfabet dan angka
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Mengganti slang dengan kata formal menggunakan kamus alay
    text = replace_slang_with_formal(text)

    # Tokenisasi dan pembersihan lebih lanjut
    words = word_tokenize(text)
    words = [word for word in words if word not in stop_words_id]
    words = [lemmatizer.lemmatize(word) for word in words]

    # Menggabungkan kembali menjadi string
    cleaned_text = ' '.join(words)

    return cleaned_text  # Pastikan kita mengembalikan hasil pembersihan

# Terapkan pembersihan teks
df['cleaned_tweet'] = df['tweet'].apply(clean_text)
```

```python

from transformers import AutoTokenizer, AutoModel
# Load model sentiment analysis yang sudah dilatih
tokenizer = AutoTokenizer.from_pretrained("w11wo/indonesian-roberta-base-sentiment-classifier")
model = AutoModelForSequenceClassification.from_pretrained("w11wo/indonesian-roberta-base-sentiment-classifier")
```

```python
# Buat pipeline untuk analisis sentimen
sentiment_model = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python

# Fungsi untuk memberikan label sentimen pada teks
def sentiment_label(text):
    # Cek jika teks kosong, jika iya, langsung return label 'neutral' atau label default
    if not text.strip():
        return 'neutral'

    # Tokenisasi input sebelum memberi ke model
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)

    # Pastikan tipe tensor yang dikirim adalah LongTensor
    inputs = {key: value.to(dtype=torch.long) for key, value in inputs.items()}

    # Mendapatkan hasil prediksi dari model
    result = model(**inputs)
    logits = result.logits
    # Mengambil label prediksi (label dengan skor tertinggi)
    predicted_class = logits.argmax(-1).item()

    # Mendapatkan label yang sesuai
    labels = sentiment_model.model.config.id2label
    label = labels[predicted_class]
    return label
```

```python
import pandas as pd
import re
df = pd.read_csv(r'/content/drive/MyDrive/tweet_sentiment_results22.csv')

def extract_mentions(tweet):
    mentions = re.findall(r'@\w+', tweet)
    return mentions

# Membuat edge list (relasi antara pengguna)
edges = []
for index, row in df.iterrows():
    tweet = row['tweet']
    mentions = extract_mentions(tweet)
    for mention in mentions:
        edges.append((row['username'], mention[1:]))  # Menghapus '@' dan menyimpan hubungan

# Menyimpan edge list dalam DataFrame
edges_df = pd.DataFrame(edges, columns=['source', 'target'])

print(edges_df.head())
```
